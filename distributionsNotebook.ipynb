```python
%matplotlib notebook
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob as glb
```

```python
def getCC4CSDataFrame(paths):
    data_frames = []
    for ele in paths:
        data_frames.append(pd.read_csv(ele, sep = ';|,', engine='python'))
    
    res = data_frames[0][' CC4CS']
    for i, ele in enumerate(data_frames[1:]):
        # print(paths[i])
        res = res.append(data_frames[i][' CC4CS'], ignore_index=True)
        
    return res


def calculateStatistics(sampleDataFrame):
    print("min: ", sampleDataFrame.max(), "max: ", sampleDataFrame.min(), "median: ", sampleDataFrame.median(), "standard deviation: ", sampleDataFrame.std())
```

```python
sns.distplot(leon8Res.dropna(), label = "int8", hist = False)
sns.distplot(leon16Res.dropna(), label = "int16", hist = False)
sns.distplot(leon32Res.dropna(), label = "int32", hist = False)
sns.distplot(floatRes.dropna(), label = "float", hist = False)
plt.legend()
plt.title("Custom Benchmark Results on Leon3")
plt.show()
```

```python
calculateStatistics(leon8Res)
calculateStatistics(leon16Res)
calculateStatistics(leon32Res)
calculateStatistics(floatRes)
```
